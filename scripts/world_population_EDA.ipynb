## cleaning missing value

```python
df.fillna(df.median(numeric_only=True), inplace=True)
df.isnull().sum()
```

## Visualization Top 10 Most Populated countries 2022

```python
top10_countries = df.nlargest(10, '2022 Population')
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='2022 Population', data=top10_countries)
plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Population Data')
plt.show()
```


## Population Growth Over Years

```python
years = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2022 Population']
top5 = df.nlargest(5, '2022 Population')
plt.figure(figsize=(12, 6))
for i, row in top5.iterrows():
    plt.plot(years, row[years], marker='o', label=row['Country'])
plt.title('Population Growth 1970-2022')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

## Top 10 Largest Countries by Area

```python
top10_area = df.nlargest(10, 'Area (km²)')

plt.figure(figsize=(12,6))
sns.barplot(x='Area (km²)', y='Country', data=top10_area)
plt.title('Top 10 Largest Countries by Area')
plt.xlabel('Area (km²)')
plt.ylabel('Country')
plt.show()
```

## Population by Continent

```python
continent_pop = df.groupby('Continent')['2022 Population'].sum()

plt.figure(figsize=(10,8))
plt.pie(continent_pop, 
        labels=continent_pop.index, 
        autopct='%1.1f%%')
plt.title('Population by Continent 2022')
plt.show()
```

## Top 10 Most Dense Countries

```python
top10_density = df.nlargest(10, 'Density (per km²)')

plt.figure(figsize=(12,6))
sns.barplot(x='Density (per km²)', y='Country', data=top10_density)
plt.title('Top 10 Most Dense Countries')
plt.xlabel('Density (per km²)')
plt.ylabel('Country')
plt.show()
```

## Outliers Detection

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[['2022 Population', 
                      'Area (km²)', 
                      'Density (per km²)']])
plt.title('Outliers Detection')
plt.show()
```

## Growth Rate Analysis

```python
top10_growth = df.nlargest(10, 'Growth Rate')

plt.figure(figsize=(12,6))
sns.barplot(x='Growth Rate', y='Country', data=top10_growth)
plt.title('Top 10 Fastest Growing Countries')
plt.xlabel('Growth Rate')
plt.ylabel('Country')
plt.show()
```

## top 10 slowest countries

```python
bottom10_growth = df.nsmallest(10, 'Growth Rate')

plt.figure(figsize=(12,6))
sns.barplot(x='Growth Rate', y='Country', data=bottom10_growth)
plt.title('Top 10 Slowest Growing Countries')
plt.xlabel('Growth Rate')
plt.ylabel('Country')
plt.show()
```
